# Kalman filter initial position estimation

To generate a trajectory we need to predict the future position of the UAV, around 40ms in the future.
We want to see if we can estimate it correctly using a Kalman filter.

Here, we plot the position estimation given by the IMU and the expected position to see if we really need a Kalman filter.

```python
import pickle, numpy as np, matplotlib.pyplot as plt
# X0 is real X and X1 is the expected X for data; X1 is the ground truth position prediction for traj generation
T, X0, Y0, Z0, X1, Y1, Z1, VX0, VY0, VZ0, VX1, VY1, VZ1 = np.arange(13) 
```

```python
data, req = np.load('pos1.npy'), np.load('req1.npy')
```

```python
plt.figure(figsize=(25,10))
plt.subplot(321)
plt.plot(data[:, T], data[:, X0], label=r'$X(t)$')
plt.plot(data[:, T], data[:, X1], label=r'$\hat{X}(t)$')
plt.scatter(req[:, T], req[:, X0])
plt.xlabel('Time')
plt.ylabel('Position (m)')
plt.legend()

plt.subplot(323)
plt.plot(data[:, T], data[:, Y0], label=r'$Y(t)$')
plt.plot(data[:, T], data[:, Y1], label=r'$\hat{Y}(t)$')
plt.xlabel('Time')
plt.ylabel('Position (m)')
plt.legend()

plt.subplot(325)
plt.plot(data[:, T], data[:, Z0], label=r'$Z(t)$')
plt.plot(data[:, T], data[:, Z1], label=r'$\hat{Z}(t)$')
plt.xlabel('Time')
plt.ylabel('Position (m)')
plt.legend()

plt.subplot(322)
plt.plot(data[:, T], data[:, VX0], label=r'$v_X(t)$')
plt.plot(data[:, T], data[:, VX1], label=r'$\hat{v_X}(t)$')
plt.xlabel('Time')
plt.ylabel('Velocity (m/s)')
plt.legend()

plt.subplot(324)
plt.plot(data[:, T], data[:, VY0], label=r'$v_Y(t)$')
plt.plot(data[:, T], data[:, VY1], label=r'$\hat{v_Y}(t)$')
plt.xlabel('Time')
plt.ylabel('Velocity (m/s)')
plt.legend()

plt.subplot(326)
plt.plot(data[:, T], data[:, VZ0], label=r'$v_Z(t)$')
plt.plot(data[:, T], data[:, VZ1], label=r'$\hat{v_Z}(t)$')
plt.xlabel('Time')
plt.ylabel('Velocity (m/s)')
plt.legend()
```
