```python
import numpy as np, matplotlib.pyplot as plt, json
```

```python
success = np.load('save1_success.npy')
failure = np.load('save1_failure.npy')
empty = np.load('save1_empty.npy')
```

```python
print('Success', success)
print('Failure', failure)
```

```python
a, b = 0, 200
plt.figure(figsize=(20,10))
plt.bar(np.arange(a, b), success[a:b], label='Success')
plt.bar(np.arange(a, b), failure[a:b], label='Failure')
plt.bar(np.arange(a, b), empty[a:b], label='Empty')
plt.legend()
plt.xlim(a, b)
plt.show()
```

```python
success_rate = np.sum(success)/(100*2*10)
print('Success rate: {}%'.format(success_rate*100))
```
