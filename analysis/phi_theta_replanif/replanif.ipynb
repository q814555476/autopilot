Replanification benchmark between Phi* and Theta*
=================================================

````python
import numpy as np, matplotlib.pyplot as plt
````

## Experiment 1

Compare Phi*/Theta* in large map with one new obstacle far away offline.
With a large obstacle. In 2D

Obstacle: 
```
bbmin = [-6, 14, 0]
bbmax = [-14, 16, 3]
```

Theta* performs better in all situations.

**Hypothesis:**
A step in Phi* is more expensive than in Theta*. But Theta* execute more steps.
So here Theta* performs better because the replanification of Phi* still takes some time, since the obstacle is large.
A smaller modification may improve greatly the performances

````python
initPhi = np.array([1.93, 2.15, 2.03, 2.16, 2.15, 2.23, 2.03, 2.15, 2.12, 2.03])
updatePhi = np.array([3.14, 3.40, 3.46, 3.67, 3.66, 3.44, 3.52, 3.36, 3.53, 3.55])

initTheta = np.array([1.36, 1.36, 1.40, 1.34, 1.45, 1.32, 1.38, 1.51, 1.43, 1.37])
updateTheta = np.array([2.70, 2.77, 2.64, 2.50, 2.74, 2.76, 2.65, 2.97, 2.88, 2.58])
````

````python
plt.figure(figsize=(10,10))
plt.boxplot([initPhi, updatePhi], positions=[.25, .5], labels=['Phi* initial', 'Phi* replan'],
            patch_artist=True, boxprops=dict(facecolor="#fdd7d7"), medianprops=dict(color="black"))

plt.boxplot([initTheta, updateTheta], positions=[.75, 1], labels=['Theta* initial', 'Theta* replan'],
            patch_artist=True, boxprops=dict(facecolor="#d2fbd2"), medianprops=dict(color="black"))
plt.ylabel('Processing time (sec)')
plt.title('Comparison Phi*/Theta* with one step of replanning, with a large obstacle close to the goal in 2D')
plt.xlim(0.15, 1.1)
#plt.savefig('time_all.png', bbox_inches='tight')
````
